Let's print out the profile for each cuisine by displaying the top four ingredients in each cuisine.

```python
num_ingredients = 10   # define number of top ingredients to print

# define a function that prints the top ingredients for each cuisine
def number_top_ingredients(row): 
    print(row.name.upper())
    row_sorted = row.sort_values(ascending = False) * 100
    top_ingredients = list(row_sorted.index.values)[0:num_ingredients]
    row_sorted = list(row_sorted)[0:num_ingredients]
    
    for ingred, ingredient in enumerate(top_ingredients):
        print("%s (%d%%)" % (ingredient, row_sorted[ingred]), end=' ')
    print("\n")

# apply function to cuisines dataframe
create_cuisines_profiles = cuisines.apply(number_top_ingredients, axis=1)
```

## [bamboo_tree] Only Asian and Indian Cuisines

Here, we are creating a decision tree for the recipes for just some of the Asian (Korean, Japanese, Chinese, Thai) and Indian cuisines. The reason for this is because the decision tree does not run well when the data is biased towards one cuisine, in this case American cuisines. One option is to exclude the American cuisines from our analysis or just build decision trees for different subsets of the data. Let's go with the latter solution.

Let's build our decision tree using the data pertaining to the Asian and Indian cuisines and name our decision tree bamboo_tree.

```python
# select subset of cuisines
asians_indians_recipes = recipes[recipes.cuisine.isin(['korean', 'japanese', 'indian', 'thai', 'chinese'])]
cuisines = asians_indians_recipes['cuisine']
ingredients = asians_indians_recipes.iloc[:,1 : ]

bamboo_tree = tree.DecisionTreeClassifier(max_depth=3)
bamboo_tree.fit(ingredients, cuisines)

print('Decision Tree model saved to Bamboo tree')
```

Let's plot the decision tree and examine how it looks like.

```python
export_graphviz(bamboo_tree,
               feature_names= list(ingredients.columns.values),
               out_file = "bamboo_tree.dot",
               class_names= np.unique(cuisines),
               filled = True,
               node_ids = True,
               special_characters=True, 
               impurity=False,
               label="all",
               leaves_parallel=False)

with open ("bamboo_tree.dot") as bamboo_tree_image:
    bamboo_tree_graph = bamboo_tree_image.read()
graphviz.Source(bamboo_tree_graph)
```

The decision tree learned:

* If a recipe contains cumin and fish and no yoghurt, then it is most likely a Thai recipe.
* If a recipe contains cumin but no fish and no soy_sauce, then it is most likely an Indian recipe.

```python
# select another subset of cuisines
asians_indians_recipes = recipes[recipes.cuisine.isin(['german', 'french', 'african', 'mexican', 'greek'])]
cuisines = asians_indians_recipes['cuisine']
ingredients = asians_indians_recipes.iloc[:,1 : ]

subset_tree = tree.DecisionTreeClassifier(max_depth=3)
subset_tree.fit(ingredients, cuisines)

print('Decision Tree model saved to Subset tree')
```

```python
export_graphviz(bamboo_tree,
               feature_names= list(ingredients.columns.values),
               out_file = "subset_tree.dot",
               class_names= np.unique(cuisines),
               filled = True,
               node_ids = True,
               special_characters=True, 
               impurity=False,
               label="all",
               leaves_parallel=False)

with open ("subset_tree.dot") as subset_tree_image:
    subset_tree_graph = subset_tree_image.read()
graphviz.Source(subset_tree_graph)
```

You can analyze the remaining branches of the tree to come up with similar rules for determining the cuisine of different recipes.

To evaluate our model of Asian and Indian cuisines, we will split our dataset into a training set and a test set. We will build the decision tree using the training set. Then, we will test the model on the test set and compare the cuisines that the model predicts to the actual cuisines.

Let's first create a new dataframe using only the data pertaining to the Asian and the Indian cuisines, and let's call the new dataframe  **sillians**.

```python
sillians = recipes[recipes.cuisine.isin(["korean", "japanese", "chinese", "thai", "indian"])]
```

Let's see how many recipes exist for each cuisine.

```python
sillians['cuisine'].value_counts()
```

Let's remove 30 recipes from each cuisine to use as the test set, and let's name this test set **sillians_test**.

```python
# set sample size
sample_n = 30
```

Create a dataframe containing 30 recipes from each cuisine, selected randomly.

```python
# take 30 recipes from each cuisine
random.seed(1234) # set random seed
sillians_test = sillians.groupby("cuisine", group_keys=False).apply(lambda x: x.sample(sample_n))

sillians_test_ingredients = sillians_test.iloc[:,1:] # ingredients
sillians_test_cuisines = sillians_test["cuisine"] # corresponding cuisines or labels
```

Check that there are 30 recipes for each cuisine.

```python
# check that we have 30 recipes from each cuisine
sillians_test["cuisine"].value_counts()
```

Next, let's create the training set by removing the test set from the **sillians** dataset, and let's call the training set **sillians_train**.



```python
sillians_test_index = sillians.index.isin(sillians_test.index)
sillians_train = sillians[~sillians_test_index]

sillians_train_ingredients = sillians_train.iloc[:,1:] # ingredients
sillians_train_cuisines = sillians_train["cuisine"] # corresponding cuisines or labels
```

Check that there are 30 fewer recipes now for each cuisine.

```python
sillians_train['cuisine'].value_counts()
```

Let's build the decision tree using the training set, **sillians_train**, and name the generated tree **sillians_train_tree** for prediction.

```python
sillians_train_tree = tree.DecisionTreeClassifier(max_depth=15)
sillians_train_tree.fit(sillians_train_ingredients, sillians_train_cuisines)

print("Decision tree model saved to sillians_train_tree!")
```

Let's plot the decision tree and explore it.

```python
export_graphviz(sillians_train_tree,
                feature_names=list(sillians_train_ingredients.columns.values),
                out_file="sillians_train_tree.dot",
                class_names=np.unique(sillians_train_cuisines),
                filled=True,
                node_ids=True,
                special_characters=True,
                impurity=False,
                label="all",
                leaves_parallel=False)

with open("sillians_train_tree.dot") as sillians_train_tree_image:
    sillians_train_tree_graph = sillians_train_tree_image.read()
graphviz.Source(sillians_train_tree_graph)
```

Now that we defined our tree to be deeper, more decision nodes are generated.

## Now let's test our model on the test data.

```python
sillians_pred_cuisines = sillians_train_tree.predict(sillians_test_ingredients)
```

To quantify how well the decision tree is able to determine the cuisine of each recipe correctly, we will create a confusion matrix which presents a nice summary on how many recipes from each cuisine are correctly classified. It also sheds some light on what cuisines are being confused with what other cuisines.

So let's go ahead and create the confusion matrix for how well the decision tree is able to correctly classify the recipes in **sillians_test.**

```python
test_cuisines = np.unique(sillians_test_cuisines)
bamboo_confusion_matrix = confusion_matrix(sillians_test_cuisines, sillians_pred_cuisines, test_cuisines)
title = 'Bamboo Confusion Matrix'
cmap = plt.cm.Blues

plt.figure(figsize=(8, 6))
bamboo_confusion_matrix = (
    bamboo_confusion_matrix.astype('float') / bamboo_confusion_matrix.sum(axis=1)[:, np.newaxis]
    ) * 100

plt.imshow(bamboo_confusion_matrix, interpolation='nearest', cmap=cmap)
plt.title(title)
plt.colorbar()
tick_marks = np.arange(len(test_cuisines))
plt.xticks(tick_marks, test_cuisines)
plt.yticks(tick_marks, test_cuisines)

fmt = '.2f'
thresh = bamboo_confusion_matrix.max() / 2.
for i, j in itertools.product(range(bamboo_confusion_matrix.shape[0]), range(bamboo_confusion_matrix.shape[1])):
    plt.text(j, i, format(bamboo_confusion_matrix[i, j], fmt),
             horizontalalignment="center",
             color="white" if bamboo_confusion_matrix[i, j] > thresh else "black")

plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')

plt.show()
```

The rows represent the actual cuisines from the dataset and the columns represent the predicted ones. Each row should sum to 100%.

Using the first row in the confusion matrix, 36.67% of the Chinese recipes in sillians_test were correctly classified by our decision tree whereas 30% of the Chinese recipes were misclassified as Korean and 20% were misclassified as Indian.

Using the Indian row, 90% of the Indian recipes in sillians_test were correctly classified by our decision tree and 3.33% of the Indian recipes were misclassified as Chinese and 0% were misclassified as Korean and 6.67% were misclassified as Thai.
